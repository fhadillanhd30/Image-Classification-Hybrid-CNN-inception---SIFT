```python
X_train_sift = []
X_test_sift = []
```

```python
for label in labels:
    train_label_path = os.path.join(train_path, label)
    test_label_path = os.path.join(test_path, label)
    for file in os.listdir(train_label_path):
        if file.endswith(".jpg"):
            img = cv2.imread(os.path.join(train_label_path, file))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img = cv2.resize(img, (224,224))
            sift = cv2.SIFT_create()
            kp, des = sift.detectAndCompute(img, None)
            kmeans = KMeans(n_clusters=1, random_state=0).fit(des)
            X_train_sift.append(kmeans.cluster_centers_)
            img = np.ravel(img)
    for file in os.listdir(test_label_path):
        if file.endswith(".jpg"):
            img = cv2.imread(os.path.join(test_label_path, file))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img = cv2.resize(img, (224,224))
            sift = cv2.SIFT_create()
            kp, des = sift.detectAndCompute(img, None)
            kmeans = KMeans(n_clusters=1, random_state=0).fit(des)
            X_test_sift.append(kmeans.cluster_centers_)
```

```python
X_train_sift = np.array(X_train_sift)
X_test_sift = np.array(X_test_sift)
```

```python
X_train_sift = np.reshape(X_train_sift,(2400,128))
X_test_sift = np.reshape(X_test_sift,(600,128))
```
